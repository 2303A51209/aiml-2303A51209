```python
import matplotlib.pyplot as plt

people  = ['gama','sushruth','aditya','sathwik','shiva','goutham','subhani']

age = [3,4,7,21,31,32,43]

weight=[30,40,50,20,65,70,30]

height=[150,155,152,160,162,170,166]

plt.scatter(weight,height)
plt.title("weight-height relation ")
plt.ylabel("height in cm")
plt.xlabel("weight in kg")
plt.show()
plt.scatter(weight,age)
plt.title("weight-age relation ")
plt.ylabel("age in years")
plt.xlabel("weight in kg")
plt.show()
```

generate bar plot

```python
plt.figure(figsize=(12,4))
plt.title("people's weight in kgs",fontsize=20,fontstyle='italic' )


plt.bar(x=people,height=weight,width=0.8,color="red" )

plt.xlabel('People',fontsize=20)
plt.ylabel("weight(in kgs)",fontsize=20)
plt.show()
```
